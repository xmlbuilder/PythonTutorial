```python
## üéØ 1. Sine & Cosine Í∑∏ÎûòÌîÑ (Matplotlib)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Îç∞Ïù¥ÌÑ∞ ÏÉùÏÑ±
x = np.linspace(0, 2 * np.pi, 100)
y_sin = np.sin(x)
y_cos = np.cos(x)

# ÏãúÍ∞ÅÌôî
plt.plot(x, y_sin, label='Sine', color='blue')
plt.plot(x, y_cos, label='Cosine', color='red')
plt.title('Sine & Cosine Functions')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# ÏãúÍ∞ÑÎ≥Ñ Í∞ÅÎèÑ Î≥ÄÌôò
hours = np.arange(0, 24)
counts = np.random.randint(10, 100, size=24)
angles = 2 * np.pi * hours / 24

# ÏãúÍ∞ÅÌôî
fig = plt.figure()
ax = fig.add_subplot(111, polar=True)
ax.plot(angles, counts, color='purple', linewidth=2)
ax.fill(angles, counts, alpha=0.3)
ax.set_title('Polar Chart - ÏãúÍ∞ÑÎ≥Ñ ÌÜµÍ≥ÑÎüâ')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 2 * np.pi, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# Ï≤´ Î≤àÏß∏ subplot (1Ìñâ 2Ïó¥ Ï§ë Ï≤´ Î≤àÏß∏)
plt.subplot(1, 2, 1)
plt.plot(x, y1, color='blue')
plt.title('Sine')

# Îëê Î≤àÏß∏ subplot (1Ìñâ 2Ïó¥ Ï§ë Îëê Î≤àÏß∏)
plt.subplot(1, 2, 2)
plt.plot(x, y2, color='red')
plt.title('Cosine')

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Îç∞Ïù¥ÌÑ∞ ÏÉùÏÑ±
x = np.linspace(0, 2 * np.pi, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# 2Ìñâ 1Ïó¥ subplot ÏÉùÏÑ±
fig, ax = plt.subplots(2, 1, figsize=(8, 6))

# Ï≤´ Î≤àÏß∏ subplot: Sine
ax[0].plot(x, y1, color='blue')
ax[0].set_title('Sine')
ax[0].set_xlabel('x')
ax[0].set_ylabel('sin(x)')
ax[0].grid(True)

# Îëê Î≤àÏß∏ subplot: Cosine
ax[1].plot(x, y2, color='red')
ax[1].set_title('Cosine')
ax[1].set_xlabel('x')
ax[1].set_ylabel('cos(x)')
ax[1].grid(True)

# Î†àÏù¥ÏïÑÏõÉ Ï†ïÎ¶¨ Î∞è Ï∂úÎ†•
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 10, 100)
y = np.exp(-x) * np.sin(2 * np.pi * x)  # ÏàòÏãù ÏßÅÏ†ë ÏûÖÎ†•

plt.plot(x, y)
plt.title(r'$y = e^{-x} \cdot \sin(2\pi x)$')
plt.xlabel(r'$x$')
plt.ylabel(r'$y$')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Îç∞Ïù¥ÌÑ∞ ÏÉùÏÑ±
x = np.linspace(0, 2 * np.pi, 100)
y_sin = np.sin(x)
y_cos = np.cos(x)

# Í∑∏ÎûòÌîÑ Ïä§ÌÉÄÏùº ÏÑ§Ï†ï (matplotlib Í∏∞Î≥∏ Ïä§ÌÉÄÏùº Ï§ë ÌïòÎÇò ÏÇ¨Ïö©)
plt.style.use('ggplot')  # Îã§Î•∏ ÏòµÏÖò: 'fivethirtyeight', 'bmh', 'classic'

# Í∑∏ÎûòÌîÑ ÌÅ¨Í∏∞ ÏÑ§Ï†ï
fig, ax = plt.subplots(figsize=(10, 6))

# Sine Í∑∏ÎûòÌîÑ: Îë•Í∑º Ï†ê ÎßàÏª§, ÍµµÏùÄ ÏÑ†
ax.plot(x, y_sin, label='Sine', color='dodgerblue',
        linewidth=2.5, marker='o', markersize=5)

# Cosine Í∑∏ÎûòÌîÑ: Î≥ÑÌëú ÎßàÏª§, Ï†êÏÑ† Ïä§ÌÉÄÏùº
ax.plot(x, y_cos, label='Cosine', color='crimson',
        linewidth=2, linestyle='--', marker='*', markersize=6)

# Ï†úÎ™© Î∞è Ï∂ï ÎùºÎ≤® (LaTeX ÏàòÏãù Ìè¨Ìï®)
ax.set_title(r'Styled Plot of $y = \sin(x)$ and $y = \cos(x)$',
             fontsize=16, fontweight='bold')
ax.set_xlabel(r'$x$ (radians)', fontsize=14)
ax.set_ylabel(r'$y$', fontsize=14)

# Î≤îÎ°Ä ÏÑ§Ï†ï
ax.legend(loc='upper right', fontsize=12, frameon=True)

# Í≤©Ïûê Ïä§ÌÉÄÏùº
ax.grid(True, linestyle=':', linewidth=0.8)

# Ï∂ï Î≤îÏúÑ ÏÑ§Ï†ï
ax.set_xlim(0, 2 * np.pi)
ax.set_ylim(-1.2, 1.2)

# Î†àÏù¥ÏïÑÏõÉ Ï†ïÎ¶¨ Î∞è Ï∂úÎ†•
plt.tight_layout()
plt.show()
```
